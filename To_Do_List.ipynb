```python
## ğŸ“Œ To-Do List Project in Python

tasks = []

# Step 2: Menu ko repeatedly show karne ke liye ek function banate hain
def show_menu():
    print("\nğŸ“Œ TO-DO LIST MENU")
    print("1. Add Task")
    print("2. View Tasks")
    print("3. Mark Task as Done")
    print("4. Delete Task")
    print("5. Exit")

# Step 3: Task add karne ka function
def add_task():
    task = input("Enter a new task: ")
    tasks.append({"task": task, "done": False})
    print(f"âœ… '{task}' added to the list.")

# Step 4: Sare tasks dikhane ka function
def view_tasks():
    if not tasks:
        print("âš ï¸ No tasks available!")
    else:
        print("\nğŸ“ Your Tasks:")
        for i, task in enumerate(tasks, 1):
            status = "âœ”ï¸ Done" if task["done"] else "âŒ Pending"
            print(f"{i}. {task['task']} [{status}]")

# Step 5: Task ko Done mark karne ka function
def mark_task_done():
    view_tasks()
    try:
        task_num = int(input("Enter task number to mark as done: "))
        tasks[task_num - 1]["done"] = True
        print("âœ… Task marked as done.")
    except (IndexError, ValueError):
        print("âš ï¸ Invalid task number.")

# Step 6: Task delete karne ka function
def delete_task():
    view_tasks()
    try:
        task_num = int(input("Enter task number to delete: "))
        removed_task = tasks.pop(task_num - 1)
        print(f"ğŸ—‘ï¸ Task '{removed_task['task']}' deleted.")
    except (IndexError, ValueError):
        print("âš ï¸ Invalid task number.")

# Step 7: Main program loop
while True:
    show_menu()
    choice = input("Enter your choice (1-5): ")

    if choice == "1":
        add_task()
    elif choice == "2":
        view_tasks()
    elif choice == "3":
        mark_task_done()
    elif choice == "4":
        delete_task()
    elif choice == "5":
        print("ğŸ‘‹ Exiting To-Do List. Goodbye!")
        break
    else:
        print("âš ï¸ Invalid choice, please try again.")
```
